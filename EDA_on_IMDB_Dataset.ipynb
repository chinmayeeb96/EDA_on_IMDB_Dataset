# 4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
# Create a dataframe which contains the non null values of crew column
imdb_df_non_na_crew = imdb_df_new.loc[imdb_df_new['crew'].notna()]
```

```python
len(imdb_df_non_na_crew)
```

```python
# Creating a fucntion which returns the list of producers and directors for each movie 
def get_producers_and_directors(row):

  try:
    # Initialize two lists
    producers_list = []
    directors_list = []

    # Iterate over the list elements
    for elem in row['crew']:
      if elem['job'] == 'Director':
        directors_list.append(elem['name'])
    
    
      if elem['job'] == 'Producer':
        producers_list.append(elem['name'])

    row['producers'] = producers_list
    row['directors'] = directors_list

    return row


  except:
    return None
                              
                              
    
```

```python
imdb_df_new2 = imdb_df_new.apply(get_producers_and_directors, axis = 1)
```

```python
imdb_df_new2.head()
```

## Alternate approach 

```python
def func_producers_and_directors(crew_list):

  try:
    # Initialize two lists
    producers_list = []
    directors_list = []

    # Iterate over the list elements
    for elem in crew_list:
      if elem['job'] == 'Director':
        directors_list.append(elem['name'])
    
    
      if elem['job'] == 'Producer':
        producers_list.append(elem['name'])

    

    return producers_list, directors_list


  except:
    return [],[]
```

```python
# Creating two columns, producers column and directors column
imdb_df_new[['producers', 'directors']] = imdb_df_new.apply(lambda row: func_producers_and_directors(row['crew']), axis = 1,result_type = 'expand')
```

```python
imdb_df_new.info()
```

```python
# Making a subset of the imdb dataframe with only roi and producers column and unstacking the list of column of producers
imdb_df_new[['producers', 'roi']].explode('producers')
```

```python
# Finding the top 3 producers with highest average roi
imdb_df_new[['producers', 'roi']].explode('producers').groupby(['producers'])['roi'].mean().sort_values(ascending=False).head(3)
```

# 5. Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
# Creating a function to get the actors list
def get_actors_list(casts_list):

  try:
    # Initialize actor list
    actors_list = []
    

    # Iterate over the list elements
    for elem in casts_list:      
      actors_list.append(elem['name'])
    
    return actors_list


  except:
    return []
```

```python
# Creating the actors column
imdb_df_new['actors'] = imdb_df_new.apply(lambda row: get_actors_list(row['cast']), axis = 1)
```

```python
imdb_df_new.head()
```

```python
# Getting the actor with highest number of movies
actor_with_movie_numbers = imdb_df_new[['original_title', 'actors']].explode('actors').groupby(['actors'])['original_title'].count().reset_index()
```

```python
actor_with_movie_numbers[actor_with_movie_numbers['original_title'] == actor_with_movie_numbers['original_title'].max()]
```

# 6. Top 3 directors prefer which actors the most?

```python
# Printing top 3 directors
top_3_directors = imdb_df_new[['directors', 'roi']].explode('directors').groupby(['directors'])['roi'].mean().sort_values(ascending=False).head(3).index.tolist()
top_3_directors
```

```python
# Creating a dataframe with the director and actor column
director_actor_df = imdb_df_new[['directors', 'actors']].explode('actors').explode('directors')
director_actor_df.head()
```

```python
# printing the actors preferred by the top 3 directors
director_actor_df[director_actor_df['directors'].isin(top_3_directors)]
```
